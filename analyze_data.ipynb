```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# CNA

```python
cna_df = pd.read_csv('data/CNA_test_data.csv')
cna_df.head(3)
```

```python
cna_df.info()
```

```python
cna_df.describe()
```

```python
# Check null
for key in cna_df.keys():
    if cna_df[key].isnull().sum() > 0:
        print(cna_df[key].isnull().sum())
```

```python
cna_df.boxplot(column=['CRC_1'], return_type='axes')
```

```python
cna_df[cna_df.CRC_1 == cna_df.CRC_1.max()]
```
